```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data from CSV file
data = pd.read_csv('StudentsPerformance.csv')
```

```python
# Assuming your dataset has 'math score' and 'reading score' columns
X = data[['math score', 'reading score']]
```

```python
# Plotting the sample data using Matplotlib
plt.scatter(X['math score'], X['reading score'])
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('Sample Data')
plt.show()
```

```python
# K-Means Clustering
def kmeans_clustering(X, n_clusters):
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)
    kmeans.fit(X)
    return kmeans.labels_
```

```python
# Example usage with 3 clusters
n_clusters = 3
kmeans = KMeans(n_clusters=n_clusters, random_state=0)
labels = kmeans.fit_predict(X)
```

```python
# Plotting clusters using Matplotlib
plt.scatter(X['math score'], X['reading score'], c=labels, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100)
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('K-Means Clustering')
plt.show()
```

```python
# Plotting clusters using Seaborn
sns.scatterplot(x='math score', y='reading score', hue=labels, data=X, palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', c='red', s=100, label='Centroids')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.title('K-Means Clustering with Seaborn')
plt.legend()
plt.show()

```
