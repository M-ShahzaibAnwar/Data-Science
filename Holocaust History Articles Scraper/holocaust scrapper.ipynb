```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
from googlesearch import search
```

```python
def scrape_article(url):
    try:
        response = requests.get(url)

        soup = BeautifulSoup(response.content, 'html.parser')

        content = soup.find('body')
        paragraphs = content.find_all('p')

        extracted_texts = []
        for para in paragraphs:
            text_content = para.get_text()  # Extract text from paragraph
            if text_content:  # Only proceed if text is not empty
                extracted_texts.append(text_content)  # Add to list




        # text = "\n".join([para.get_text() for para in paragraphs if para.get_text()])

        # return text

    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return None
```

```python
query = "Holocaust history articles"

max_articles = 100

articles = []

count = 0
```

```python
for url in search(query):
    if count >= max_articles:
        break

    print(f"Scraping {url}...")

    article_text = scrape_article(url)
    if article_text:
        articles.append({"url": url, "text": article_text})
        count += 1
    time.sleep(1)
```

```python
df = pd.DataFrame(articles)
df.to_csv('Holocaust_ART.csv', index=False)
print("Scraping complete. Data saved to Holocaust_ART.csv.")
```
