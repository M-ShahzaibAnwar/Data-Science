```python
from googlesearch import search
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time
```

```python
def scrape_article(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')


        content = soup.find('body')
        paragraphs = content.find_all('p')

        text = "\n".join([para.get_text() for para in paragraphs if para.get_text()])

        return text
    except Exception as e:
        print(f"Failed to scrape {url}: {e}")
        return None
```

```python
# List to hold article texts
articles = []
```

```python
query = "Holocaust history articles"
max_articles = 50
count = 0

for url in search(query):
    if count >= max_articles:
        break

    print(f"Scraping {url}...")
    article_text = scrape_article(url)

    if article_text:
        articles.append({"url": url, "text": article_text})
        count += 1  # Increment the counter for each successful scrape

    time.sleep(1)  # To prevent overloading the server

# Convert the list of articles into a DataFrame
df = pd.DataFrame(articles)

# Save the DataFrame to a CSV file
df.to_csv('holocaust_articles_dynamic.csv', index=False)

print("Scraping complete. Data saved to holocaust_articles_dynamic.csv.")

```
