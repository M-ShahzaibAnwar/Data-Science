```python
# SHAHZAIB ANWAR S22-BCS-054
# ROLL NO SP22-BCS-054
# make dictionary whith some predefined data of followers saved values on lists
thisdict = {
    "followers" : ["10k", "1000k", "1k", "100k", "20k","200k", "900k", "25k", 500, "11k", "12k",
                   "13k", "14k", "15k", "16k", "17k", "18k", "19k", "20k", "21k"],
    # initalize some empty lists
    "users" : [],
    "plateform" : [],
    "likes_per_user" : [],
    "views" : [],
    "posts" : []
}
```

```python
# take input from the user
n=int(input("how many time you wanna input"))
# loop n times
for i in range(n):
  username = input("Enter your username: ")
  plateform = input("Enter your plateform: ")
  likes_per_user = (input("Enter your likes per user: "))
  views = (input("Enter your views: "))
  posts = (input("Enter your posts: "))
# use append method to attach at the end
  thisdict["users"].append(username)
  thisdict["plateform"].append(plateform)
  thisdict["likes_per_user"].append(likes_per_user)
  thisdict["views"].append(views)
  thisdict["posts"].append(posts)


```

```python
# convert dictionaries into dataframes
import pandas as pd
df=pd.DataFrame(thisdict)
print(df)
```

```python
n=int(input("how many time you wanna update"))
for i in range(n):
  # Ask user which row (index) they want to update
  row_idx = int(input("\nEnter the row index (0-19) you want to update: "))

  # Ask user which column they want to update
  col_name = input(f"Enter the column to update {list(df.columns)}: ")

  # Ask for the new value
  new_value = input(f"Enter the new value for {col_name}: ")

  # Perform the update
  df.at[row_idx, col_name] = new_value

  print(f"\nAfter updating {col_name} for user at row {row_idx}:")
  print(df)
```

```python
# Define the new row data
new_followers = "100k"
new_user = "ali"
new_plateform = "instagram"
new_likes_per_user = 7
new_views = "8000"
new_posts = 7

# Create a new row DataFrame
new_row = pd.DataFrame({
    "followers": [new_followers],
    "users": [new_user],
    "plateform": [new_plateform],
    "likes_per_user": [new_likes_per_user],
    "views": [new_views],
    "posts": [new_posts],
})
new_index=len(df)
# Insert the new row at the end of the DataFrame
df = pd.concat([df, new_row], ignore_index=True)

# Print the DataFrame after insertion
print("\nAfter inserting new row:")
print(df)
```

```python
index_to_delete = 22 # You can change this to delete other users by index

# Delete the row at the specified index
df = df.drop(index_to_delete)

# Reset the index if necessary
df.reset_index(drop=True, inplace=True)

# Print the DataFrame after deletion
print("\nAfter deleting user at index 22:")
print(df)
```
