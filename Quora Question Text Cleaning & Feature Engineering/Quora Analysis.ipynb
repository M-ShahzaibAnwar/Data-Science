```python
data = df.loc[:, ~df.columns.str.contains('^Unnamed')]
```

```python
data.head()
```

```python
data=data.dropna(subset=["question1","question2","is_duplicate"])
```

```python
data.duplicated().sum()
```

```python
data.columns
```

```python
data['question1'] = data['question1'].str.lower()
data['question2'] = data['question2'].str.lower()
```

```python
data.head()
```

```python
import re
```

```python
def clean_text(text):
  #removal of special characters

  text = re.sub(r"[^a-zA-Z\s]", "", text)

  #removal of multiple spaces
  text = re.sub(r'\s+', ' ', text)
  #removal of leading space
  text=text.strip()
  return text

data['question1']=data['question1'].apply(clean_text)
data['question2']=data['question2'].apply(clean_text)
```

```python
data.head()
```

```python
from nltk.stem import PorterStemmer


stemmer = PorterStemmer()

def stem_text(text):

    return ' '.join([stemmer.stem(word) for word in text.split()])

data['question1_stemmed'] = data['question1'].apply(stem_text)
data['question2_stemmed'] = data['question2'].apply(stem_text)

print(data[['question1', 'question1_stemmed']].head())
print(data[['question2', 'question2_stemmed']].head())
```

```python
from nltk.stem import WordNetLemmatizer
import nltk
nltk.download('wordnet')
nltk.download('omw-1.4')

lemmatizer = WordNetLemmatizer()
```

```python
# Function to apply lemmatization
def lemmatize_text(text):
    return ' '.join([lemmatizer.lemmatize(word) for word in text.split()])

# Apply lemmatization to the cleaned text
data['question1_lemmatized'] = data['question1'].apply(lemmatize_text)
data['question2_lemmatized'] = data['question2'].apply(lemmatize_text)

# View the results
print(data[['question1', 'question1_lemmatized']].head(20))
data[['question2', 'question2_lemmatized']].head()
```

```python
def remove_duplicates(text):
    words = text.split()
    result = []
    for word in words:
        if word not in result:
            result.append(word)
    return ' '.join(result)
```

```python
data['question1_no_duplicates'] = data['question1_lemmatized'].apply(remove_duplicates)
data['question2_no_duplicates'] = data['question2_lemmatized'].apply(remove_duplicates)

print(data[['question1_lemmatized', 'question1_no_duplicates', 'question2_lemmatized', 'question2_no_duplicates']].head())
```

```python
data['q1_length'] = data['question1_no_duplicates'].apply(lambda x: len(str(x)))
data['q2_length'] = data['question2_no_duplicates'].apply(lambda x: len(str(x)))
```

```python
data['common_words'] = data.apply(
    lambda x: len(set(str(x['question1_no_duplicates']).split()).intersection(set(str(x['question2_no_duplicates']).split()))), axis=1)
```

```python
data['same_start'] = data.apply(
    lambda x: str(x['question1_no_duplicates']).split()[0] == str(x['question2_no_duplicates']).split()[0]
    if str(x['question1_no_duplicates']).split() and str(x['question2_no_duplicates']).split()
    else False,
    axis=1
)
```
